## Загрузка датасета 2007 TREC Public Spam Corpus
1. Прочитайте и приймите соглашение"Agreement for use" по ссылке https://plg.uwaterloo.ca/~gvcormac/treccorpus07/

2. Затем загрузите корпус "255 MB Corpus (trec07p.tgz)" и разархивируйте в директорию './datasets'

3. Убедитесь, что используемые ниже пути 'DATA_DIR' 'LABELS_FILE' существуют.

```python
DATA_DIR = 'datasets/trec07p/data/'
LABELS_FILE = 'datasets/trec07p/full/index'
TRAINING_SET_RATIO = 0.7
TESTING_SET_RATIO = 1 - TRAINING_SET_RATIO
```

```python
labels = {}
# Получаем метки классов
with open(LABELS_FILE) as f:
    for line in f:
        line = line.strip()
        label, key = line.split()
        labels[key.split('/')[-1]] = 1 if label.lower() == 'ham' else 0
```

```python
def read_email_files():
    X = []
    y = [] 
    for i in range(len(labels)):
        filename = 'inmail.' + str(i+1)
        email_str = email_read_util.extract_email_text(
            os.path.join(DATA_DIR, filename))
        X.append(email_str)
        y.append(labels[filename])
    return X, y
```

```python
X, y = read_email_files()
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test, idx_train, idx_test = \
    train_test_split(X, y, range(len(y)), 
    train_size = TRAINING_SET_RATIO, test_size = TESTING_SET_RATIO, random_state = 2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_vector = vectorizer.fit_transform(X_train)
X_test_vector = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# Инициализация классификатора и предсказание классов
mnb = MultinomialNB()
mnb.fit(X_train_vector, y_train)
y_pred = mnb.predict(X_test_vector)

# Вывод результатов
print(classification_report(y_test, y_pred, target_names=['Spam', 'Ham']))
print('Classification accuracy {:.1%}'.format(accuracy_score(y_test, y_pred)))
```
